```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.read_csv('missing_value_free.csv')
data.drop(columns=['Unnamed: 0','CPU Name','GPU Name','OS'],axis=1,inplace=True)
data
```

```python
import pickle
import pickle
def encode(brand): 
    LabelEncoder_brands = LabelEncoder()
    LabelEncoder_brands.fit(data['Brand'])
    return LabelEncoder_brands.transform([brand])[0]
data['Encoded_Brand'] = data['Brand'].apply(encode)
price_column = data.pop('Price')
data['Price'] = price_column
encoded_brand_dict = dict(zip(data['Brand'],data['Encoded_Brand']))
# Specify the file path to save the dictionary
file_path = 'encoded_brand_dict.pkl'

# Save the dictionary to the file
with open(file_path, 'wb') as file:
    pickle.dump(encoded_brand_dict, file)
    # Specify the file path to load the dictionary
file_path = 'encoded_brand_dict.pkl'

print(len(encoded_brand_dict))

print(encoded_brand_dict)
data
```
